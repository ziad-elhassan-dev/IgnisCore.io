```python
import json
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
from aiBrain.FireDetector import FireDetector
```

```python
with open("simulated_fire.json") as f:
    data = json.load(f)

detector = FireDetector()
preprocessed = detector.preprocess(data)
```

```python
scores = detector.calculate_fire_risk(preprocessed)
action = detector.detect_fire(scores)
detector.log_result(scores, action)
```

```python
df = pd.DataFrame([scores])
df.plot(kind='bar', y=['temp','smoke','ir','global'], title="Fire Risk Scores")
plt.ylabel("Score")
plt.xlabel("Metrics")
plt.ylim(0, 1)
plt.show()
```
